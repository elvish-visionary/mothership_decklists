```python
import pandas as pd
import numpy as np

df = pd.read_csv('legacy_top_8.csv')
df.head()
```

```python
matrix = df.pivot_table(index='Winning deck', columns='Losing deck', values='Date', aggfunc='count')
matrix.fillna(0, inplace=True)
for deck in list(matrix.columns):
    try:
        print("{} - Wins: {}, Losses: {}".format(deck, matrix.loc[deck,:].sum(), matrix.loc[:, deck].sum()))
    except KeyError:
        print("{} - Wins: 0, Losses: {}".format(deck, matrix.loc[:, deck].sum()))
```

```python
wincounts = pd.DataFrame(df[df['Winning deck'] != df['Losing deck']].groupby('Winning deck').size())
losecounts = pd.DataFrame(df[df['Winning deck'] != df['Losing deck']].groupby('Losing deck').size())
wincounts.index.rename('Deck', inplace=True)
losecounts.index.rename('Deck', inplace=True)
wincounts.columns = ['Wins']
losecounts.columns = ['Losses']


wincounts.merge(losecounts, how='left', left_index=True, right_index=True).to_csv('top8winslosses.csv')
```
